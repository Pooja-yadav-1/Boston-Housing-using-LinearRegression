```python
#Importing libraries of python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
```

```python
#Splits data in Training and Testing sets
from sklearn.model_selection import train_test_split

X=boston.drop('medv',axis=1)
Y=boston['medv']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape )
```

```python
train_data=X_train.join(Y_train)
train_data.head()
```

```python
#Creation of Histogram of Train dataset
train_data.hist(color='green', edgecolor='purple')
```

```python
#Graphical representaion of individual data
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="ocean")
```

```python
#Finding Intercept, Slope and Root Mean Square Error Using Linear Regression
X0 = boston.iloc[:,1:14].values
Y0 = boston.iloc[:,-1].values
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
LR0 = LinearRegression()
LR0.fit(X0,Y0)
print("Intercept : ", LR0.intercept_)
print("Slope : ", LR0.coef_)
Y0_pred = LR0.predict(X0)
RMSE0 = np.sqrt(mean_squared_error(Y0,Y0_pred))
print("Root Mean Square Error : ", RMSE0)
```

```python
#Plotting Data with Linear Regression
fig, axs = plt.subplots(nrows=2,ncols=4,figsize=(20,8))
cols = ['lstat','rm','indus','ptratio','tax','nox','crim','age']
for col,ax in zip(cols,axs.flat):
    sns.regplot(x=boston[col],y=boston['medv'],color = 'purple',ax=ax)
```

```python
#Collapsing multi-dimensional array into a one-dimensional array
fig, axs = plt.subplots(nrows=5,ncols=3,figsize=(10,8))
index = 0
axs = axs.flatten()
for k,v in boston.items():
    sns.boxplot(y=k,data=boston,ax=axs[index], color='yellow')
    index+=1
plt.tight_layout()
```

```python
boston=boston[~((boston['crim']>=25))]
boston=boston[~((boston['rm']>=8.5)|(boston['rm']<4))]
boston=boston[~((boston['lstat']>=30))]
```

```python
X_DF = boston[['lstat','rm','indus','ptratio','tax','nox','crim','age']]
X = X_DF.values
X
```

```python
Y_DF = boston[['medv']]
Y = Y_DF.values
Y
```

```python
LR = LinearRegression()
LR.fit(X,Y)
```

```python
#Printing Intercept and Slope Values
print("Intercept : ", LR.intercept_)
print("Slope : ", LR.coef_)
```

```python
#Predicted Values
Y_pred = LR.predict(X)
print("Predicted Values :\n",Y_pred)
```

```python
#Evaluating Root Mean Square Error
RMSE = np.sqrt(mean_squared_error(Y,Y_pred))
print("Root Mean Square Error : ", RMSE)
```
